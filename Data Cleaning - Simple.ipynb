```python
import pandas as pd

# Messy data

df = pd.read_csv(r'C:\Users\nicks\Documents\Electronics.csv', encoding='latin1')

print("Original DataFrame:")
print()
print(df.head())

# 1. Clean 'Product' column (lowercase and remove duplicates)
df['Product'] = df['Product'].str.lower().str.strip()
df.drop_duplicates(subset=['Product', 'Price'], inplace=True) # Drop duplicates based on Product and Price

# 2. Convert 'Price' to numeric, coercing errors to NaN
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# 3. Handle missing 'Quantity' values (e.g., fill with median)
median_quantity = df['Quantity'].median()
df['Quantity'].fillna(median_quantity, inplace=True)

# 4. Standardize 'Category'
df['Category'].replace('Elec.', 'Electronics', inplace=True)

print("\nCleaned DataFrame:")
print()
print(df.head())
```
